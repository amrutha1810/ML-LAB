```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
data = yf.download(tickers, start="2024-01-01", end="2024-12-30", group_by='ticker')
print("First 5 rows of the dataset:")
print(data.head())
# Step 3: Basic Data Exploration
# Check the shape of the dataset
print("\nShape of the dataset:")
print(data.shape)

# Check column names
print("\nColumn names:")
print(data.columns)


```

```python
hdfc_data = data['HDFCBANK.NS']
print("\nSummary statistics for HDFC Industries:")
print(hdfc_data.describe())

# Calculate daily returns
hdfc_data['Daily Return'] = hdfc_data['Close'].pct_change()

plt.subplot(2, 1, 1)
hdfc_data['Close'].plot(title="HDFC Industries - Closing Price")
plt.subplot(2, 1, 2)
hdfc_data['Daily Return'].plot(title="HDFC Industries - Daily Returns", color='orange')
plt.tight_layout()
plt.show()
# Step 4: Saving the Processed Data to a New CSV File
# Save the Reliance data to a CSV file
hdfc_data.to_csv('hdfc_data.csv')
print("\nHDFC stock data saved to 'hdfc_data.csv'.")
```

```python
ICIC_data = data['ICICIBANK.NS']
print("\nSummary statistics for ICIC Industries:")
print(ICIC_data.describe())

# Calculate daily returns
ICIC_data['Daily Return'] = ICIC_data['Close'].pct_change()

plt.subplot(2, 1, 1)
ICIC_data['Close'].plot(title="ICIC Industries - Closing Price")
plt.subplot(2, 1, 2)
ICIC_data['Daily Return'].plot(title="ICIC Industries - Daily Returns", color='orange')
plt.tight_layout()
plt.show()
# Step 4: Saving the Processed Data to a New CSV File
# Save the Reliance data to a CSV file
ICIC_data.to_csv('ICIC_data.csv')
print("\nICIC stock data saved to 'ICIC_data.csv'.")
```
